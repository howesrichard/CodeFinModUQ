```python
# use the official-joke-api to get a random joke

import pandas as pd
import random
import requests

# use the official-joke-api to get a random joke
url = 'https://official-joke-api.appspot.com/random_joke'
response = requests.get(url)
joke = response.json()
print(joke['setup'])
print(joke['punchline'])

# put the response into a pandas dataframe
df = pd.DataFrame(joke, index=[0])
```

```python
# use the api.open-meteo.com to get the current weather for a user provided latitude and longitude
latitude = -34.89
longitude = 151.19
url = f'https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true&hourly=temperature_2m,relativehumidity_2m,windspeed_10m'
response = requests.get(url)
weather = response.json()
print(weather)

# print current_weather
print(weather['current_weather'])
```

```python
# use textblob to get the sentiment of "my mum is great"
from textblob import TextBlob

text = 'my mum is great'
blob = TextBlob(text)
print(blob.sentiment)
```
